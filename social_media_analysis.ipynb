```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('smmh.csv')
```

## What is the average age of respondents?

```python
import matplotlib.pyplot as plt

# Compute the average age
average_age = df["1. What is your age?"].mean()

# Display the result
print(f"The average age of respondents is {average_age:.2f} years.")
    
```

**Conclusion:** The average age of respondents is 26.14.

## What is the gender distribution of respondents?

```python
gender_counts = df["2. Gender"].value_counts()

plt.figure(figsize=(8, 4))
gender_counts.plot(kind='bar', color=['hotpink', 'pink'])
plt.title("Gender Distribution of Respondents")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Display the result
gender_counts
```

**Conclusion:** The gender distribution is visualized in a bar chart.

## What is the most common relationship status among respondents?

```python
relationship_counts = df["3. Relationship Status"].value_counts()

plt.figure(figsize=(6, 4))
relationship_counts.plot(kind='bar', color='brown')
plt.title("Relationship Status Distribution")
plt.xlabel("Relationship Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Display the most common status
relationship_counts
    
```

**Conclusion:** The most common relationship status is determined and visualized in graph.

## What is the most common occupation status?

```python
occupation_counts = df["4. Occupation Status"].value_counts()

plt.figure(figsize=(6, 4))
occupation_counts.plot(kind='bar', color='brown')
plt.title("Occupation Status Distribution")
plt.xlabel("Occupation Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Display the most common occupation
occupation_counts
    
```

**Conclusion:** The most common occupation status is University Student.

## How many respondents use social media?

```python

# Compute social media usage count
social_media_counts = df["6. Do you use social media?"].value_counts()

# Display the count
social_media_counts
    
```

**Conclusion:** The number of respondents who use social media is determined and 478 respondents are using Social Media.

## Which social media platforms are most commonly used?

```python
platforms = df["7. What social media platforms do you commonly use?"].str.split(", ").explode()

platform_counts = platforms.value_counts()

plt.figure(figsize=(8, 4))
platform_counts.plot(kind='bar', color='brown')
plt.title("Most Commonly Used Social Media Platforms")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Display the counts
platform_counts
    
```

**Conclusion:** The most commonly used social media platforms are YouTube, Facebook and Instagram.

## What is the average daily time spent on social media?

```python

# Convert time spent to numerical values
time_map = {
    "Less than 1 hour": 0.5,
    "Between 1 and 2 hours": 1.5,
    "Between 2 and 3 hours": 2.5,
    "Between 3 and 4 hours": 3.5,
    "Between 4 and 5 hours": 4.5,
    "More than 5 hours": 6
}
df["Time Spent (Hours)"] = df["8. What is the average time you spend on social media every day?"].map(time_map)

# Compute the average time spent
average_time_spent = df["Time Spent (Hours)"].mean()

# Display the result
print(f"The average daily time spent on social media is {average_time_spent:.2f} hours.")
    
```

**Conclusion:** The average daily time spent on social media is 3.76.

## How often do respondents use social media without a specific purpose?

```python
usage_purpose_counts = df["9. How often do you find yourself using Social media without a specific purpose?"].value_counts().sort_index()

plt.figure(figsize=(6, 4))
usage_purpose_counts.plot(kind='bar', color='brown')
plt.title("Usage of Social Media Without a Specific Purpose")
plt.xlabel("Frequency (Scale 1-5)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Display the result
usage_purpose_counts
    
```

**Conclusion:** The frequency of using social media without a specific purpose is 4(163).

## How frequently do respondents get distracted by social media?

```python
distraction_counts = df["10. How often do you get distracted by Social media when you are busy doing something?"].value_counts().sort_index()

plt.figure(figsize=(6, 4))
distraction_counts.plot(kind='bar', color='brown')
plt.title("Distraction by Social Media Frequency")
plt.xlabel("Frequency (Scale 1-5)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Display the result
distraction_counts
    
```

**Conclusion:** The frequency of social media distraction is analyzed.

```python
restlessness_counts = df["11. Do you feel restless if you haven't used Social media in a while?"].value_counts().sort_index()

plt.figure(figsize=(6, 4))
restlessness_counts.plot(kind='bar', color='brown')
plt.title("Restlessness When Not Using Social Media")
plt.xlabel("Restlessness Level (Scale 1-5)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Display the result
restlessness_counts
    
```
